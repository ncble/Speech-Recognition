```python
%matplotlib nbagg
import matplotlib.pyplot as plt
import numpy as np
%pylab
from __future__ import print_function, division
# needed for plotting, imports also numpy:
```

```python
def f_sphere(x):
    if len(x.shape) ==1:
        # For 1 sample
        return sum(np.power(x,2))
    else:
        # For N samples
        return np.sum(np.power(x,2), axis = 1)
def f_step(x):
    return -f_sphere(x.astype(int))
```

```python
def draw_surface_level(fun, centre = np.zeros(2), taille = 1.0, message = None):

    plt.figure(figsize=(5, 3))
    axes = plt.gca()
    if message:
        plt.title(message)
    x_min, y_min = (centre - taille)#[0,0], (centre - taille)[0,1]  #X[:, 0].min()
    x_max, y_max = (centre + taille)#[0,0], (centre + taille)[0,1]     #X[:, 0].max()
    
    axes.set_xlim([x_min,x_max])
    axes.set_ylim([y_min,y_max])
    XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j]

    Z = fun(np.c_[XX.ravel(), YY.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(XX.shape)
#     plt.pcolormesh(XX, YY, Z, cmap = plt.cm.jet)
    CS = plt.contour(XX, YY, Z, cmap = plt.cm.jet)
    plt.clabel(CS, fmt='%2.1f', colors='b', fontsize=14)
    plt.show()
```

```python
draw_surface_level(f_step, taille = 10., message = "Step")
```
